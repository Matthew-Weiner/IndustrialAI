```python
# Define the sampling rate (in Hz) and duration (in seconds)
sampling_rate = 2560
duration = 15

# Calculate the number of samples
num_samples = int(sampling_rate * duration)

# Generate the time array
time_array = np.arange(0, duration, 1/sampling_rate)
```

```python
# This seems useless.....

# Define a function to plot time-domain and frequency-domain data
def plot_time(data_frames, title):
    plt.figure(figsize=(12, 6))
    # Plot time-domain data
    i = 0
    for df in data_frames:
        i = i+ 1
        data = df.iloc[5:, 0]
        plt.plot(time_array, data.astype(float), label= f"{i}")
        plt.title(f'{title} - Time Domain')
        plt.xlabel('Time (s)')
        plt.ylabel('Acceleration')
        plt.legend()
        plt.ylim(-0.85, 0.85)
    plt.show()

# Plot time domain for H's and U's
plot_time(H_data_frames, 'Healthy Accelerometer Data')
plot_time(U_data_frames, 'Unhealthy Accelerometer Data')
```

```python
MaxHealth = []
MaxUnhealth = []
    
def plot_freq(data_frames, title, Max):
    i = 0
    for df in data_frames:
        i = i+1
        accel = df.iloc[5:, 0]
        accel = accel.astype(float)
        time = time_array
        fft_values = np.fft.fft(accel)
        freq = np.fft.fftfreq(accel.size, d=1/sampling_rate)
        freqs_magnitude = np.abs(fft_values) / (accel.size/2)
        freq_axis = np.linspace(0, sampling_rate, len(freqs_magnitude))
        plt.plot(freq, freqs_magnitude, label = f"{i}")
        
        
        
        # Find the indices corresponding to the frequency range [20, 30] Hz
        freq_range_indices = np.where((freq >= 20) & (freq <= 30))[0]

        # Find the maximum amplitude within the specified frequency range
        max_amplitude = np.max(freqs_magnitude[freq_range_indices])
        Max.append(max_amplitude)
        
        
    plt.title(f'{title} - Frequency Domain (FFT)')
    plt.xlabel('Frequency (Hz)')
    plt.xlim(0,60)
    plt.ylim(0, .005)
    plt.ylabel('Amplitude')
    plt.legend(loc = "right")
    plt.show()
                 
plot_freq(H_data_frames, 'Healthy Accelerometer Data', MaxHealth)
plot_freq(U_data_frames, 'Unhealthy Accelerometer Data', MaxUnhealth)
```

```python
# Create a list of indices from 1 to the length of MaxHealth (or MaxUnhealth)
indices = list(range(1, len(MaxHealth) + 1))

# Plot MaxHealth and MaxUnhealth on the same graph
plt.plot(indices, MaxHealth, label='Max Healthy', marker='o')
plt.plot(indices, MaxUnhealth, label='Max Unhealthy', marker='o')

# Label the axes and add a legend
plt.xlabel('Sample')
plt.ylabel('Max Amplitude')
plt.title('Max Amplitude Comparison')
plt.legend()

# Show the plot
plt.show()

print(MaxHealth)
print(MaxUnhealth)
```
